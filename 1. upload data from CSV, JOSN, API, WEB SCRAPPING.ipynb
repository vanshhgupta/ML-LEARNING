```python
def rename(name):
    if name=="male" or "Male":
        return "M"
    elif name== None:
        return None
    else:
        return name

rename('male')

pd.read_csv("aug_train.csv",index_col='enrollee_id', converters={'gender':rename}, nrows=100, na_values=['Nan'])
```
